## Cleaning Data

```python
## Cleaning of the tests dataset

# Keeping only the columns that is interests us
clean_test_timeseries = test_timeseries[["state_name", "date", "new_results_reported"]]

# Summing the neg and pos tests altogether for a given date, state 
clean_test_timeseries = clean_test_timeseries.groupby(["state_name", "date"]).sum().reset_index()

# Renaming the columns appropriatly
clean_test_timeseries.rename({"state_name" : "State", "date" :"Date", "new_results_reported" : "TestsReported"}, axis = 1, inplace = True)

# Converting the Date in DateTime
clean_test_timeseries.Date = pd.to_datetime(clean_test_timeseries.Date)

# Removing state that are not present in the other Dataset
clean_test_timeseries = clean_test_timeseries[~clean_test_timeseries.State.isin(["Marshall Islands", "Northern Mariana Islands", "U.S. Virgin Islands", "Guam"])]

# Plotting final dataset
clean_test_timeseries
```

```python
## Cleaning of the state dataset

# Keeping only the columns that is interests us
clean_data_states = data_states[["province_state", "reported_date", "population", "confirmed", "deaths", "recovered", "people_hospitalized", "hospitalization_rate", "incident_rate", "mortality_rate", "testing_rate"]]

# Renaming the columns appropriatly
clean_data_states.rename({"province_state" : "State", "reported_date" :"Date"}, axis = 1, inplace = True)

# Converting the Date in DateTime
clean_data_states.Date = pd.to_datetime(clean_data_states.Date)

# Removing state that are not present in the other Dataset
clean_data_states = clean_data_states[~clean_data_states.State.isin(["Grand Princess", "Diamond Princess", "American Samoa", "Recovered", "Guam", "Virgin Islands", "Northern Mariana Islands"])]

# Filling NaN with 0
clean_data_states.recovered = clean_data_states.recovered.fillna(0)
clean_data_states.people_hospitalized = clean_data_states.people_hospitalized.fillna(0)
clean_data_states.hospitalization_rate = clean_data_states.hospitalization_rate.fillna(0)

# Plotting final dataset
clean_data_states
```

```python
# Merging

final_df = clean_data_states.merge(clean_test_timeseries, how = "left", on = ["State", "Date"])
final_df.sort_values(["State", "Date"], inplace = True)
final_df
```

```python
# to csv

final_df.to_csv("data/fnal_df.csv")
```
